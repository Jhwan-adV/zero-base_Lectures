#### 2. Add Geometric Information - Functions

```python
# Function: getting lat, lng info from dataframe's ['address'] column

def AddGeometryInfo(DataFrame):
    lat_list = []; lng_list = []
    for s in tqdm_notebook(DataFrame['Address']):
        try:
            g = gmaps.geocode(s)[0]
            lat = g['geometry']['location']['lat']
            lng = g['geometry']['location']['lng']
            lat_list.append(lat)
            lng_list.append(lng)
        except: # Some Address cannot get lat, lng. In that case, fill in the list with enpty str ''
            lat_list.append('')
            lng_list.append('')
            
    DataFrame['lat'] = lat_list
    DataFrame['lng'] = lng_list

    return DataFrame
```

```python
# Function: return indexes which cannot call lat, lng information 

def getInavailAddr(DataFrame):
    idx = list(filter(lambda x: DataFrame['lat'][x] == '', range(len(DataFrame))))
    if len(idx) != 0:
        return idx
    else:
        print('No empty lat, lng')
```

```python
# Fucnction: change the address into right one
    # return 4 words in the address (도로명주소 4번쨰 어절)

def AddrChange(DataFrame, idx):
    for i in idx:
        old = DataFrame['Address'][i]
        new = ' '.join(old.split()[:4]).strip(',') # strip for a single address with ','
        DataFrame.loc[i, 'Address'] = new
```

```python
# Function: update lng, lat info of address changed.

def UpdateGeocode(DataFrame, idx):
    for i in idx:
        g = gmaps.geocode(DataFrame.loc[i, 'Address'])[0]
        lat = g['geometry']['location']['lat']
        lng = g['geometry']['location']['lng']
        
        DataFrame.loc[i, 'lat'] = lat
        DataFrame.loc[i, 'lng'] = lng
```

### 3 - (4) Data Visualization - Folium

#### 1. Functions & Variables

```python
# Function: get lat, lng info from dataframe

def getLat(geocode):
    return geocode[0]['geometry']['location']['lat']

def getLng(geocode):
    return geocode[0]['geometry']['location']['lng']
```

```python
# Function: generate a map
    # set "zoom_start = 13" to get a good view on each district

def GenerateMap(lat, lng):
    m = folium.Map(
        location = [lat, lng],
        zoom_start  = 13,
        tiles='CartoDB positron',
        width = 750,
        height = 750
    )
    return m
```

```python
# Function: Add markers and cicles on the map

def EdiyaMarkers(DataFrame, Area, Map):
    df = DataFrame[DataFrame['Area'] == Area]
    for idx, rows in df.iterrows():            
        folium.Marker(
            [rows['lat'], rows['lng']],
            icon=folium.Icon(
                color='blue',
                icon='cloud'),
            tooltip = '이디야 ' + rows['Branch'], # Name '이디야'
            popup = folium.Popup(' '.join(rows['Address'].split(' ')[:3]),  max_width=1000),
        ).add_to(Map)
    
def StarbucksMarkers(DataFrame, Area, Map):
    df = DataFrame[DataFrame['Area'] == Area]
    for idx, rows in df.iterrows():            
        folium.Marker(
            [rows['lat'], rows['lng']],
            icon=folium.Icon(
                color='green',
                icon='star'),
            tooltip = '스타벅스 ' + rows['Branch'], # Name '스타벅스'
            popup = folium.Popup(' '.join(rows['Address'].split(' ')[:3]), max_width=1000),
        ).add_to(Map)

def StarbucksCircle(DataFrame, Area, Map):
    df = DataFrame[DataFrame['Area'] == Area]
    for idx, rows in df.iterrows():            
        folium.Circle(
            [rows['lat'], rows['lng']],
            radius=100,
            color = 'green', # 반경 100m 
            tooltip = '스타벅스 ' + rows['Branch'],
            popup = folium.Popup(' '.join(rows['Address'].split(' ')[:3]), max_width=1000),
        ).add_to(Map)
```

#### Method 2. Address and strategies

도로주소명의 구분 "대로 > 로 > (길, 번길)"에 따라 각 브랜드의 입점 비율을 분석합니다.

```python
# Function - count ro and gil

def CntRoGil(DataFrame):
    list_ro = []; list_gil = []; list_dong = []
    for idx, row in DataFrame.iterrows():
        addr = row['Address']
        if addr.split(' ')[2][-1] == '로':
            list_ro.append(addr) 
        elif addr.split(' ')[2][-1] == '길':
            list_gil.append(addr)
        else:
            list_dong.append(addr)
    
    return [len(list_ro), len(list_gil), len(list_dong)]
```

```python
# Functions - calculate the ratio

def CalRoGil(DataFrame):
    tmp = CntRoGil(DataFrame)

    hub_spoke = round(tmp[0] / sum(tmp) * 100, 2)
    point_point = round(tmp[1] / sum(tmp) * 100, 2)
    return [hub_spoke, point_point]
```

```python
# Perform - Entire Seoul

sb_rat = CalRoGil(starbucks)
print(f'스타벅스 서울 전 지점 1. 대로/로 입점 비율: {sb_rat[0]}%, 2. 길/번길 입점 비율: {sb_rat[1]}%')

ed_rat = CalRoGil(ediya)
print(f'이디야 서울 전 지점 1. 대로/로 입점 비율: {ed_rat[0]}%, 2. 길/번길 입점 비율: {ed_rat[1]}%')
```

서울시 전체를 대상으로 한 두 브랜드의 주소별 대로/로 입점비율 차이는 14.63%를, 길/번길 입점비율 차이는 11.87%를 보입니다. 

```python
# Perform - By Area

rate_dic = {}
for area in areas:
    tmp_sb = starbucks[starbucks['Area'] == area]
    tmp_ed = ediya[ediya['Area'] == area]
    sb_rat = CalRoGil(tmp_sb)
    ed_rat = CalRoGil(tmp_ed)

    rate_dic[area] = [sb_rat[0], ed_rat[0], sb_rat[1], ed_rat[1]]
```

```python
df_rate['Starbucks Hub'] = [v[0] for v in list(rate_dic.values())]
df_rate['Ediya Hub'] = [v[1] for v in list(rate_dic.values())]
df_rate['Starbucks PtoP'] = [v[2] for v in list(rate_dic.values())]
df_rate['Ediya PtoP'] = [v[3] for v in list(rate_dic.values())]

for idx, row in df_rate.iterrows():
    df_rate.loc[idx, 'gap'] = abs(row['Starbucks Hub'] - row['Ediya Hub'])

df_rate.head()
```

그 중에서 스타벅스의 대로/로 입점비율과의 차이가 5% 미만인 곳은 2개 구 입니다. (동작구, 종로구)
