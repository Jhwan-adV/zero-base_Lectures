```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rc
rc('font', family = 'KoPubWorldDotum_Pro')
```

```python
oil = pd.read_csv('2/data/02_oil_prices.csv', encoding='cp949')
```

```python
oil.head()
oil.set_index('index')
```

```python
oil['LPG_YN'] = 'N'
oil.head()
```

```python
def drawGasoline(df):
    plt.figure(figsize=(20,12))
    sns.boxplot(x='Area', y='Gasoline_P', data = df, hue = 'SELF_YN', palette = 'rocket')
    plt.grid
    plt.show

drawGasoline(oil)
```

```python
def drawDiesel(df):
    plt.figure(figsize=(20,12))
    sns.boxplot(x='Area', y='Diesel_P', data = df, hue = 'SELF_YN', palette = 'Paired')
    plt.grid
    plt.show

drawDiesel(oil)
```
